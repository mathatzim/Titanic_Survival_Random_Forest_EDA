```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plotting style
sns.set(style='whitegrid')
```

```python
# Step 2: Load the dataset
from google.colab import files
uploaded = files.upload()
```

```python
# Replace 'train.csv' with the uploaded filename if needed
df = pd.read_csv("train.csv")
```

```python
# Step 3: Basic dataset information
print("DataFrame Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Step 4: Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Visualizing missing data helps identify incomplete columns
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Step 5: Univariate Analysis
# Shows overall survival distribution
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
# Distribution of passenger ages, useful to see skewness and outliers
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
# Counts of passengers in each class (1st, 2nd, 3rd)
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Step 6: Bivariate Analysis
# Compares survival by sex  shows that more females survived
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.show()
# Survival by class  more 1st class passengers survived
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
# Age comparison between survivors and non-survivors
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.show()
# Relationship between Age and Fare, colored by survival
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare by Survival')
plt.show()
```

```python
# Step 7: Correlation Heatmap
# Shows strength of relationships between numeric variables
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
